Цель: на данной практической работе необходимо использовать датасет с прошлой практической работы (StudentsPerformance) и визуализировать данные в виде графиков.

Этот код выполняет несколько действий:
1. Импортирует библиотеку для визуализации данных matplotlib.pyplot. Библиотека numpy для работы с массивами данных. Библиотека pandas для работы с данными. Библиотека seaborn для визуализации данных.
2. Считывает данные из файла StudentsPerformance.csv и сохраняет их в переменную data в формате датасет. После выполнения этого кода, данные из файла StudentsPerformance.csv будут загружены в переменную data и будут готовы для анализа и визуализации.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas  as pd
import seaborn as sns

data = pd.read_csv("/Users/kseniapankova/Desktop/StudentsPerformance.csv")

data
```

Этот код используется для создания и отображения графика, используя библиотеку визуализации данных matplotlib. Создается новый график c заданным размером 20x10 дюймов. график строиться, используя данные из столбца 'math score' для оси x и данные из столбца 'reading score' для оси y, но только для последних 5 строк (.tail(5)), и делает это в виде линии. Также задается метка для этой линии с помощью параметра label. Ось x подписана как "Оценка по математике", а ось y подписана как "Оценка по чтению". Заголовок графика "Статистика оценок за математику и чтение" 

```python
plt.figure(figsize=(20,10))
plt.plot(data['math score'].tail(5),data['reading score'].tail(5), label = 'Оценки')
plt.title('Статистика оценок за математику и чтение')
plt.xlabel('Оценка по математике')
plt.ylabel('Оценка по чтению')
plt.legend()
plt.show()
```

Этот код создает график с оценками за чтение для последних пяти записей в столбце "gender". 
Первая строка устанавливает размер графика 20 на 10 дюймов. Затем на графике строится линия, отображающая оценки за чтение для последних пяти записей, используя данные из столбца "gender" в качестве оси x и из столбца "reading score" в качестве оси y. Линия отмечена как "Оценки за чтение" с розовым цветом, а точки на графике обозначены маркерами в форме кругов с размером 7. Заголовок графика устанавливается как "Статистика оценок за чтение", а подписи осей x и y устанавливаются как "Пол" и "Оценка за чтение" соответственно. Также добавлена легенда и сетка для улучшения читаемости графика. 

```python
plt.figure(figsize=(20,10))
plt.plot(data['gender'].tail(5),data['reading score'].tail(5), label = 'Оценки за чтение', color = "pink", marker='o', markersize=7)
plt.title('Статистика оценок за чтение')
plt.xlabel('Пол')
plt.ylabel('Оценка за чтение')
plt.legend()
plt.grid()
plt.show()
```

Этот код создает круговую диаграмму, отображающую соотношение между наличием и отсутствием подготовки к экзамену. Функция value_counts() подсчитывает количество уникальных значений в столбце 'test preparation course'. Затем plt.pie() строит круговую диаграмму на основе этих значений, где каждый сектор диаграммы представляет собой одно уникальное значение, а его размер соответствует количеству этого значения в данных. Заголовок диаграммы устанавливается с помощью plt.title(). Из анализа графика можно сделать выводы о том, что преобладает в данных и какова их относительная частота. Это может быть полезной информацией для понимания сколько учащихся готовяться к экзамену.

```python
test = data['test preparation course'].value_counts()
plt.pie(test, labels=test.index, colors = ['purple','yellow'])
plt.title('Соотношение подготовки к экзамену')
plt.show()
```

Этот код создает круговую диаграмму, отображающую соотношение уровня образования родителей учащихся. Функция value_counts() подсчитывает количество уникальных значений в столбце 'parental level of education'. Затем plt.pie() строит круговую диаграмму на основе этих значений, где каждый сектор диаграммы представляет собой одно уникальное значение, а его размер соответствует количеству этого значения в данных. Значения в секторах диаграммы подписаны с использованием autopct. Из анализа графика можно сделать выводы о том, какое образование у родителей учащихся. Это может быть полезной информацией для понимания какое образование у родителей учащихся. 

```python
a = data['parental level of education'].value_counts()
plt.pie(a, labels=a.index, autopct='%1.1f%%', colors = ['orange','pink','purple','brown','red','blue'])
plt.title('Соотношение уровня образования родителей')
plt.show()
```

Этот код создает диаграмму рассеяния, отображающую связь между полом и расой/этнической принадлежностью для первых 5 записей в данных. Функция plt.figure(figsize=(7,4)) устанавливает размеры фигуры, на которой будет отображаться диаграмма. Затем plt.scatter() строит диаграмму рассеяния, где каждая точка представляет собой комбинацию пола и расы/этнической принадлежности для одного учащегося. Цвет точек установлен в фиолетовый с помощью параметра color = 'purple'. Заголовок диаграммы устанавливается с помощью plt.title(), а метки осей x и y устанавливаются с помощью plt.xlabel() и plt.ylabel(). Наконец, plt.show() отображает созданную диаграмму.

```python
plt.figure(figsize=(7,4))
plt.scatter(data['gender'].head(5), data['race/ethnicity'].head(5), color = 'purple')
plt.title('Связь между полом и расой/этнической принадлежностью')
plt.xlabel('Пол')
plt.ylabel('Раса/этническая принадлежность')
plt.show()
```

Этот код создает диаграмму рассеяния, отображающую связь между оценкой за написанную работу и оценкой за чтение для данных в датасете. Функция plt.figure(figsize=(10, 5)) устанавливает размеры фигуры, на которой будет отображаться диаграмма. Затем plt.scatter() строит диаграмму рассеяния, где каждая точка представляет собой комбинацию оценки за написанную работу и оценки за чтение для одного учащегося. Наконец, plt.show() отображает созданную диаграмму.

```python
plt.figure(figsize=(10, 5))
plt.scatter(data['writing score'], data['reading score'], color = 'blue')
plt.title('Связь между оценкой за написанную работу и оценку за чтение')
plt.xlabel('Оценка за написанную работу')
plt.ylabel('Оценка за чтение')
plt.show()
```

Этот код создает гистограмму оценок за чтение (reading score) из данных в датасете. Функция plt.figure(figsize=(20,10)) устанавливает размеры фигуры, на которой будет отображаться гистограмма. Затем plt.hist() строит гистограмму, показывая распределение оценок за чтение. Цвет столбцов установлен в фиолетовый (color = "purple"), цвет контуров столбцов - в желтый (edgecolor = 'yellow').
```python
plt.figure(figsize=(20,10))
plt.hist(data['reading score'], color = "purple", edgecolor = 'yellow', label = 'Пол')
plt.legend()
plt.show()
```

Этот код создает гистограмму оценок по математике из данных в датасете. Функция plt.figure(figsize=(18,10)) устанавливает размеры фигуры, на которой будет отображаться гистограмма. Затем plt.hist() строит гистограмму, показывая распределение оценок по математике. Цвет столбцов установлен в розовый (color = "pink"), цвет контуров столбцов - в красный (edgecolor = 'red'). Подпись "Оценки по математике" устанавливается с помощью параметра label = 'Оценки по математике', а затем с помощью plt.legend() добавляется легенда.
```python
plt.figure(figsize=(18,10))
plt.hist(data['math score'], bins=30, color = ["pink"], edgecolor = 'red', label = 'Оценки по математике')
plt.legend()
plt.show()
```

Этот код создает гистограмму оценок за письменную работу из данных в датасете. Функция plt.figure(figsize=(18,10)) устанавливает размеры фигуры, на которой будет отображаться гистограмма. Затем plt.hist() строит гистограмму, показывая распределение оценок за письменную работу. Цвет столбцов установлен в коричневый (color = "brown"), цвет контуров столбцов - в черный (edgecolor = 'black'). Подпись "Оценки за письменную работу" устанавливается с помощью параметра label = 'Оценки за письменную работу', а затем с помощью plt.legend() добавляется легенда.
```python
plt.figure(figsize=(18,10))
plt.hist(data['writing score'], bins=30, color = ["brown"], edgecolor = 'black', label = 'Оценки за письменную работу')
plt.legend()
plt.show()
```
